```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import time
```

```python
place = []
country_city = []
description = []
link = []

for i in range(1, 12):
    time.sleep(3)
    url = 'https://www.viajerosporelmundo.com/explora-el-mundo/pagina/%s'
    r = requests.get(url % i).content
        
    soup = BeautifulSoup(r)
    pla = soup.select('div.texto h2')
    places = [item.text for item in pla]
    coun = soup.select('div.info-single span.localizacion')
    countries = [item.text for item in coun]
    desc = soup.select('div.texto p')
    descriptions = [item.text for item in desc]
    links = [a['href'] for a in soup.select('div.texto h2 a')]
    
    place.append(places)
    country_city.append(countries)
    description.append(descriptions)
    link.append(links)
```

```python
placeDF = pd.Series(place)
#placeDF
complete_place = placeDF.explode()
#complete_place


country_cityDF = pd.Series(country_city)
#country_cityDF
complete_country_city = country_cityDF.explode()
#complete_country_city


descriptionDF = pd.Series(description)
#descriptionDF
complete_description = descriptionDF.explode()
#complete_description


linkDF = pd.Series(link)
#linkDF
complete_link = linkDF.explode()
#complete_link
```

```python
information_from_urls = []

for i in complete_link:
    info = requests.get(i).content
        
    soup2 = BeautifulSoup(info)
    information = soup2.select('div #bloque-archivo p')
    information_from_urls.append(information)
```

```python
information_from_urls
```

```python
complete_information = pd.Series(information_from_urls)
complete_information
```

```python
dataset = pd.concat([complete_place, complete_country_city, complete_description, complete_link], axis=1)
dataset
```

```python
ViajerosPorElMundo = dataset
ViajerosPorElMundo.columns = ['destino_turistico', 'pais-ciudad', 'descripcion', 'link']
ViajerosPorElMundo
```

```python
ViajerosPorElMundo.to_csv('viajeros_por_el_mundo.csv', index=False)
```
